### line Plots 

```python
x= [1,2,3,4,5]
y= [10,20,25,30 , 50]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.grid(True)
```

### Bar Chart

```python
months = ['jan' , 'feb' , 'march' , 'april']
revenue = [100 , 200 , 300 , 1000]
```

```python
plt.figure(figsize = (10 , 6))
plt.bar(months , revenue);
plt.title('Monthly revenue bar chart')
plt.ylabel('Revenue')
plt.xlabel('Months')
```

### Subplots and legends

```python
days = [1,2,3,4,5]
original_price = [10,20,30,40,50]
predicted_price = [8 , 34 , 22 , 67 , 11]
```

```python
plt.plot(days , original_price , 'b-o' , label = 'Original Price')
plt.plot(days , predicted_price , 'r-o', label = 'Predicted Price')
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
```

### subplots

```python
plt.subplot(3,1,1)
plt.plot(days , original_price , 'b-o' , label = 'Original Price')
plt.title("Original Prices")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

plt.subplot(3,1,2)
plt.plot(days , predicted_price , 'r-o', label = 'Predicted Price')
plt.title("Predicted Prices")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()


price_differnce = [  o - p for o , p in zip(original_price ,  predicted_price  )]
plt.subplot(3,1,3)
plt.plot(days , predicted_price ,  color = 'purple', label = 'Price difference')
plt.title("Prices diffrences")
plt.xlabel("days")
plt.ylabel("Price Differnce")
plt.legend()
plt.tight_layout()
```
